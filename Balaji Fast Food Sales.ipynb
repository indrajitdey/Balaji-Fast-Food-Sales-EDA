# Data Cleaning

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['transaction_type']=df['transaction_type'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
from datetime import datetime

# Example dates in different formats
dates = df['date']

# Convert each date to the desired format
formatted_dates = []

for date_str in dates:
    # Use multiple format strings to try parsing the date
    try:
        date_object = datetime.strptime(date_str, "%m-%d-%Y")
    except ValueError:
        try:
            date_object = datetime.strptime(date_str, "%m/%d/%Y")
        except ValueError:
            print(f"Cannot parse date: {date_str}")
            continue
    
    # Format the datetime object to the desired format
    formatted_date = date_object.strftime("%Y-%m-%d")
    formatted_dates.append(formatted_date)

# Print the formatted dates
# for formatted_date in formatted_dates:
#     print(formatted_date)
    
```

```python
df['date']=formatted_dates
```

```python
df.sample(20)
```

```python
df
```

```python
df['received_by'].unique()
```

```python
for gen in df['received_by']:
    if gen=='Mr.':
        df['received_by'].replace('Mr.','Male',inplace=True)
    elif gen=='Mrs.':
        df['received_by'].replace('Mrs.','Female',inplace=True)
```

```python
df['received_by']
```

```python
df.rename(columns = {'received_by':'Gender'}, inplace = True) 
```

```python
df.sample(5)
```

## 1. Business Growth by Time

```python
sales_by_date=df.groupby('date')['transaction_amount'].sum().reset_index()
sales_by_date
```

```python
sns.lineplot(x="date", y="transaction_amount", data=sales_by_date)
# changing the size of figure to 2X2
plt.figure(figsize=(30, 2))
# display(plt.plot(x, y))
plt.show()
```

## 2. Sales by Gender

```python
Sales_by_gender=df.groupby('Gender')['transaction_amount'].sum().sort_values(ascending=False).reset_index()
Sales_by_gender
```

```python
plt.pie(x=Sales_by_gender['transaction_amount'],labels=Sales_by_gender['Gender'])
```

## 3. Item Type

```python
item=df.groupby('item_type')['quantity'].sum().sort_values(ascending=False).reset_index()
item
```

```python
plt.pie(x=item['quantity'],labels=item['item_type'])
```

# 4. top 3 items

```python
top_3_item=df.groupby('item_name')['transaction_amount'].sum().sort_values(ascending=False)[:3].reset_index()
top_3_item
```

```python
sns.barplot(x=top_3_item['item_name'],y=top_3_item['transaction_amount'])
```

## 5. Time

```python
sales_time=df.groupby('time_of_sale')['transaction_amount'].sum().sort_values(ascending=False).reset_index()
sales_time
```

```python
sns.barplot(x='time_of_sale',y='transaction_amount',data=sales_time)
```

## 6. Transaction Type

```python
transaction_type=df.groupby('transaction_type')['order_id'].count().sort_values(ascending=False).reset_index()
transaction_type
```

```python
plt.pie(labels=transaction_type['transaction_type'],x=transaction_type['order_id'],autopct='%1.02f%%')
```
